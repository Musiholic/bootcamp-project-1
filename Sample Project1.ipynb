```python
# Create a clean DataFrame by dropping the duplicate Artist by their song.
clean_music_chart=music_chart.drop_duplicates(subset=['Artist Name', 'Song Name'])
clean_music_chart.head()
```

```python
# Create a clean DataFrame by dropping the duplicate Artist by their lyrics.
clean_music_lyrics=music_lyrics.drop_duplicates(subset=['Artist Name', 'Song Name'])
clean_music_lyrics.head()
```

```python
# Merging the music chart data with the music lyrics data
music_chart_complete=pd.merge(clean_music_chart, clean_music_lyrics[["Artist Name", "Song Name","Lyrics"]],how="left", on=["Artist Name", "Song Name"])
display(music_chart_complete)
```

```python
# Determine correlated factors of the Days and Top 10(xTimes) of the music chart. 
plt.scatter(music_chart_complete.iloc[:,3],music_chart_complete.iloc[:,4])
plt.title('Correlation')
plt.xlabel('Days')
plt.ylabel('Top 10(xTimes)')
plt.savefig("name")
plt.show()
```

```python
# The next example will compute the Pearson correlation coefficient between "Days" and "Top 10 (xTimes)"
days = music_chart_complete.iloc[:,3]
top_10 = music_chart_complete.iloc[:,4]
correlation = st.pearsonr(days,top_10)
print(f"The correlation between both factors is {round(correlation[0],2)}")

# Analysis; The correlation value of 0.57 for the Days against the Top 10 xTimes of the Music chart means its positive moderate correlation
```

```python
# Make a scatter plot to show the Position and Total Streams of the Music chart

plt.scatter(music_chart_complete.iloc[:,0],music_chart_complete.iloc[:,8]
plt.xlabel('Position')
plt.ylabel('Total Streams')
plt.show()

```

```python
## The next example will compute the Pearson correlation coefficient between Position and Total Streams

position = music_chart_complete.iloc[:,0]
total_streams = music_chart_complete.iloc[:,8]
correlation = st.pearsonr(position,total_streams)
plt.scatter(position,total_streams)
plt.title('Correlation')
plt.xlabel('Position')
plt.ylabel('Total Streams')
print(f"The correlation between both factors is {round(correlation[0],2)}")
plt.show()


# Analysis; The correlation value of -0.51 for the Position against the Top Total Streams of the Music chart means its negative moderate correlation
```

```python
# Print out the r-squared value along with the Position and Total Streams plot.
x_values =  music_chart_complete['Position']
y_values =  music_chart_complete['Total Streams']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6.0,0.9),fontsize=15,color="purple")
plt.title('Regression')
plt.xlabel('Position')
plt.ylabel('Total Streams')
print(f"The r-squared is: {rvalue**2}")
plt.show()
```

```python
# Print out the r-squared value along with the Days and Peak Streams plot.
x_values =  music_chart_complete['Days']
y_values =  music_chart_complete['Peak Streams']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5.6,0.7),fontsize=16,color="purple")
plt.title('Regression')
plt.xlabel('Days')
plt.ylabel('Peak Sreams')
print(f"The r-squared is: {rvalue**2}")
plt.savefig("name")
plt.show()
```
