```python
# Import dependencies
import requests
import pandas as pd
from azapi import AZlyrics
import time
```

```python
# Import dataset
data = pd.read_csv("top_10k_streamed_songs.csv")

# Narrow dataset to artist and song
data = data[["Artist Name","Song Name"]]

# Add new column for lyrics
data[["Lyrics"]] = ""

# Display
data
```

```python
data = pd.read_csv("Lyrics.csv")
data = data[["Artist Name","Song Name","Lyrics"]]
```

```python
# Loop through dataset and retrieve lyrics for each entry
api = AZlyrics()
for i in range(len(data)):

    try:


        if data.iloc[i,2] == "":
            api.artist = data.iloc[i,0]
            api.title = data.iloc[i,1]
            response = api.getLyrics()

            data.iloc[i,2] = response
            print(f"Adding {api.title}. Track {i} of {len(data)}.")
            time.sleep(5)
    except IndexError:
        print("Skipping")
        pass



```
